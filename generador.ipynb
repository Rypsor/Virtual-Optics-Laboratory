```python
from PIL import Image
import matplotlib.pyplot as plt

def floyd_steinberg_halftone(image_path, output_path):
    """
    Aplica el tramado por difusiÃ³n de error de Floyd-Steinberg a una imagen.
    """
    try:
        img = Image.open(image_path).convert('L')
    except FileNotFoundError:
        print(f"Error: No se encontrÃ³ el archivo en '{image_path}'")
        return

    # Usa el mÃ©todo 'dither' con el algoritmo FLOYDSTEINBERG
    binary_img = img.convert('1', dither=Image.FLOYDSTEINBERG)

    try:
        binary_img.save(output_path)
        print(f"âœ… Tramado Floyd-Steinberg guardado en '{output_path}'")
    except Exception as e:
        print(f"Error al guardar: {e}")

# --- Ejemplo de uso ---
input_image = "images/pedro2.png" # Puedes usar una imagen a color, se convertirÃ¡ a escala de grises internamente
output_image_fs = "images/generadas/imagen1_fs.png"
floyd_steinberg_halftone(input_image, output_image_fs)

# Mostrar resultado
plt.imshow(Image.open(output_image_fs), cmap='gray')
plt.title("Tramado Floyd-Steinberg")
plt.axis('off')
plt.show()
```
